```python
#7.
#Merge all four steps of the funnel, in order, using a series of left merges. Save the results to the variable all_data.

#Examine the result using print and head
all_data = pd.merge(visit, cart, how='left').merge(checkout, how='left').merge(purchase, how='left')
all_data
```

```python
import plotly.graph_objects as go

# Load your data (assuming it's already loaded as all_data)

# Calculate the counts for each step
total_visits = len(all_data)
cart_adds = all_data['cart_time'].notna().sum()
checkouts = all_data['checkout_time'].notna().sum()
purchases = all_data['purchase_time'].notna().sum()

# Create a funnel chart
stages = ["Visit", "Cart Add", "Checkout", "Purchase"]
values = [total_visits, cart_adds, checkouts, purchases]

fig = go.Figure(go.Funnel(
    y=stages,
    x=values,
    textinfo="value+percent initial"))

fig.update_layout(title="User Journey Funnel Analysis")
fig.show()
```

**9.**
Which step of the funnel is weakest (i.e., has the highest percentage of users not completing it)?

How might Cool T-Shirts Inc. change their website to fix this problem?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data (assuming it's already loaded as all_data)

# Calculate the counts for each step
total_visits = len(all_data)
cart_adds = all_data['cart_time'].notna().sum()
checkouts = all_data['checkout_time'].notna().sum()
purchases = all_data['purchase_time'].notna().sum()

# Calculate drop-off percentages
no_cart = ((total_visits - cart_adds) / total_visits) * 100
cart_to_checkout_drop = ((cart_adds - checkouts) / cart_adds) * 100 if cart_adds > 0 else 0
checkout_to_purchase_drop = ((checkouts - purchases) / checkouts) * 100 if checkouts > 0 else 0

# Data for the bar chart
stages = ["No Cart", "Cart to Checkout Drop", "Checkout to Purchase Drop"]
drop_off_rates = [no_cart, cart_to_checkout_drop, checkout_to_purchase_drop]

# Create bar chart
plt.figure(figsize=(8, 5))
plt.bar(stages, drop_off_rates, color=['red', 'orange', 'blue'])

# Add percentage labels
for i, v in enumerate(drop_off_rates):
    plt.text(i, v + 1, f"{v:.2f}%", ha='center', fontsize=12, fontweight='bold')

# Labels and title
plt.ylabel("Drop-off Percentage (%)")
plt.xlabel("Stages")
plt.title("User Drop-off Rates at Each Stage")
plt.ylim(0, 100)

plt.show()
```
